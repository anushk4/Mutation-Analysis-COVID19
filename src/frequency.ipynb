Here, we use the dataframe that was created in `main.ipynb` to plot a graphical representation.

The dataframe has two columns: `Gene` and `Frequency`.

We plot a graph of the genes against the number of times they mutate across all the strains that were taken from different countries while comparing to China.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gene_mutation_frequency.csv')
```

From the below bar chart, we interpret from our taken dataset that `ORF1ab` is most prone to mutation and `ORF6` is the most conserved domain. However, since the dataset is small, it cannot be taken as a conclusion of this chart. Similar methods can be used to find out the conserved genes for larger datasets, which can prove to be useful for drug development.

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#aec7e8']

plt.figure(figsize=(8, 6))
bars = plt.bar(df['Gene'], df['Frequency'], color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), va='bottom', ha='center')

plt.xlabel('Genes')
plt.ylabel('Mutation Frequencies')
plt.title('Mutation Frequencies of different COVID-19 Genes')
plt.savefig('../plots/mutation_frequencies.jpg')
plt.show()
```
